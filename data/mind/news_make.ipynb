```python
import pandas as pd
import numpy as np
```

```python
news_id=pd.read_csv('../mind/newid_user_id_news.csv')
news_id.head()
```

```python
#将news列名改名为id
news_id.rename(columns={'news':'id'},inplace=True)
news_id.head()
```

```python
#读取新闻数据
news_ta=pd.read_table("../mind/news_click.tsv",
              header=None,
              names=[
                  'id', 'category', 'subcategory', 'title', 'abstract', 'url',
                  'title_entities', 'abstract_entities'
              ])
news_ta.head()
```

```python
#删除category和subcategory title_entities abstract_entities
news_ta.drop(['category', 'subcategory', 'title_entities', 'abstract_entities'],axis=1,inplace=True)
news_ta.head()
```

```python
#合并新闻数据和用户数据
news_user=pd.merge(news_id,news_ta,on='id')
```

```python
news_user.head()
```

```python
news_user.shape
```

```python
#将id重复的数据删除
news_user.drop_duplicates(subset=['id'],keep='first',inplace=True)
news_user.shape
```

```python
news_user.head()
```

```python
#按照news_new排序
news_user.sort_values(by='news_new',inplace=True)
news_user.head()
```

```python
news_user.tail()
```

```python
#删掉user_id列，ave_time列,user_id_new列
news_user.drop(['user_id','ave_time','user_id_new'],axis=1,inplace=True)
news_user.head()
```

```python
#保存数据
news_user.to_csv('news_info_sorted.csv',index=False)
```
