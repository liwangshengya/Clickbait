```python
#从neww_user_item_dict_t中取出每个用户交互的项目的10%(如果交互项目数大于2至少取出1个)作为测试集，剩下的作为训练集
import random
train_user_item_dict={}
test_user_item_dict={}
for user in new_user_item_dict_t:
    #
    if len(new_user_item_dict_t[user])>2:
        test_num=int(len(new_user_item_dict_t[user])*0.35)
   #如果交互项目数小于等于2，则测试集取出1个
    if len(new_user_item_dict_t[user])==2:
        test_num=1
    #如果交互项目数小于等于1，则测试集取出0个
    if len(new_user_item_dict_t[user])<=1:
        test_num=0
    #从该用户的交互项目中随机取出test_num个作为测试集
    test_item=random.sample(new_user_item_dict_t[user],test_num)
    #将测试集中的项目从该用户的交互项目中删除
    train_item=new_user_item_dict_t[user]-set(test_item)
    #将该用户的训练集和测试集保存到字典中
    train_user_item_dict[user]=train_item
    test_user_item_dict[user]=test_item
test_user_item_dict


```

```python
#将train_user_item_dict和test_user_item_dict 的项目列表转为set
train_user_item_dict_set={}
for user in train_user_item_dict:
    train_user_item_dict_set[user]=set(train_user_item_dict[user])
test_user_item_dict_set={}
for user in test_user_item_dict:
    test_user_item_dict_set[user]=set(test_user_item_dict[user])
```

```python
#找出出现在test中但是没有出现在train中的项目
test_item=set()
train_item=set()
for user in test_user_item_dict:
    test_item=test_item.union(test_user_item_dict[user])
for user in train_user_item_dict:
    train_item=train_item.union(train_user_item_dict[user])
test_no_train_item=test_item-train_item
test_no_train_item
```

```python
len(test_item)
```

```python
len(train_item)
```

```python
#将test_no_train_item中的项目从test_user_item_dict中删除,并添加到train_user_item_dict中
for user in test_user_item_dict_set:
    no_item=test_user_item_dict_set[user]-test_no_train_item
    test_user_item_dict_set[user]=test_user_item_dict_set[user]-no_item
    train_user_item_dict_set[user]=train_user_item_dict_set[user].union(no_item)
    #如果test_user_item_dict[user]为空，则从train_user_item_dict中选出一个项目放入test_user_item_dict[user]
    if len(test_user_item_dict_set[user])==0:
       for item in train_user_item_dict_set[user]:
            if item not in test_no_train_item:
                test_user_item_dict_set[user].add(item)
                train_user_item_dict_set[user].remove(item)
                break
test_user_item_dict_set
```

```python
#从train_user_item_dict_set中取出每个用户交互的项目的10%(如果交互项目数大于2至少取出1个)作为验证集，剩下的作为训练集
import random
train_user_item_dict={}
valid_user_item_dict={}
for user in train_user_item_dict_set:
    #
    if len(train_user_item_dict_set[user])>2:
        valid_num=int(len(train_user_item_dict_set[user])*0.4)
   #如果交互项目数小于等于2，则测试集取出1个
    if len(train_user_item_dict_set[user])==2:
        valid_num=1
    #如果交互项目数小于等于1，则测试集取出0个
    if len(train_user_item_dict_set[user])<=1:
        valid_num=0
    #从该用户的交互项目中随机取出test_num个作为测试集
    valid_item=random.sample(train_user_item_dict_set[user],valid_num)
    #将测试集中的项目从该用户的交互项目中删除
    train_item=train_user_item_dict_set[user]-set(valid_item)
    #将该用户的训练集和测试集保存到字典中
    train_user_item_dict[user]=train_item
    valid_user_item_dict[user]=valid_item
```

```python
#将train_user_item_dict和test_user_item_dict 的项目列表转为set
train_user_item_dict_set={}
for user in train_user_item_dict:
    train_user_item_dict_set[user]=set(train_user_item_dict[user])
valid_user_item_dict_set={}
for user in valid_user_item_dict:
    valid_user_item_dict_set[user]=set(valid_user_item_dict[user])
```

```python
#找出出现在test中但是没有出现在train中的项目
valid_item=set()
train_item=set()
for user in valid_user_item_dict:
    valid_item=valid_item.union(valid_user_item_dict[user])
for user in train_user_item_dict:
    train_item=train_item.union(train_user_item_dict[user])
valid_no_train_item=valid_item-train_item
```

```python
#将valid_no_train_item中的项目从valid_user_item_dict中删除,并添加到train_user_item_dict中
for user in valid_user_item_dict_set:
    no_item=valid_user_item_dict_set[user]-valid_no_train_item
    valid_user_item_dict_set[user]=valid_user_item_dict_set[user]-no_item
    train_user_item_dict_set[user]=train_user_item_dict_set[user].union(no_item)
    #如果valid_user_item_dict[user]为空，则从train_user_item_dict中选出一个项目放入valid_user_item_dict[user]
    if len(valid_user_item_dict_set[user])==0:
       for item in train_user_item_dict_set[user]:
            if item not in valid_no_train_item:
                valid_user_item_dict_set[user].add(item)
                train_user_item_dict_set[user].remove(item)
                break
valid_user_item_dict_set
```

```python
#从user_item_dict中去掉test_user_item_dict_set和valid_user_item_dict_set中的用户

train_user_item_dict_set_final={}
for user in train_user_item_dict_set:
    train_user_item_dict_set_final[user]=user_item_dict[user]-test_user_item_dict_set[user]-valid_user_item_dict_set[user]
train_user_item_dict_set_final
```

```python
#计算test_user_item_dict_set中总的交互数目
test_num=0
for user in test_user_item_dict_set:
    test_num=test_num+len(test_user_item_dict_set[user])
test_num
```

```python
#计算train_user_item_dict_set_final中总的交互数目
train_num=0
for user in train_user_item_dict_set_final:
    train_num=train_num+len(train_user_item_dict_set_final[user])
train_num
```

```python
#计算valid_user_item_dict_set中总的交互数目
valid_num=0
for user in valid_user_item_dict_set:
    valid_num=valid_num+len(valid_user_item_dict_set[user])
valid_num
```
